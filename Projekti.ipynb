# Project
## Business Understanding
This project looks at a dataset about student achievement in secondary education. The attributes include grades, demographic, social and school related features. The purpose of this project is to predict whether students will get a passing grade and to find out which features affect grades the most. This can help schools decide what to use their resources on and to target students who might need additional resources based on these features. This project focuses on finding out how well you can predict student performance with information any school or school district should have easy access to.

#### Research questions
- Can you predict passing grades with the whole dataset
- Can you do it while limiting the data to features the school should have readily available
- Which features affect students passing the most

## Data Understanding
For this project the dataset from [https://archive.ics.uci.edu/dataset/320/student+performance](https://archive.ics.uci.edu/dataset/320/student+performance) is used which contains a dataset with 649 instances with 30 features. There are no missing values in the dataset.  
2 diffent versions of the model will be built. One that only includes variables that a school should already have in their system or be otherwise readily available. The second model will use more features and will be used to compare to and evaluate the accuracy of the primary model.
#### Features in both Models
sex - student's sex (binary: 'M' / 'F')  
age - student's age (integer: between 15 to 22)  
address - students home address type (binary: 'R' rural / 'U' urban)  
famsize - family size (binary: 'LE3' less or equal to 3 / 'GE3' greater than 3)  
Pstatus - parent's cohabitation status (binary: 'T' together / 'A' apart)  
guardian - student's guardian (nominal: 'mother', 'father' or 'other)  
traveltime - home to school travel time (numeric: 1 = <15 min, 2 = 15 to 30 min, 3 = 30 to 60 min, 4 = >60 min)  
failures - number of past class failures (numeric: 1 to 4, 4 if over 4)  
schoolsup - extra educational support (binary: 'yes' or 'no')  
paid - extra paid classes for the subject (binary: 'yes' or 'no')  
nursery - attended nursery school (binary: 'yes' or 'no')  
absences - number of school absences (numeric: from 0 to 93)  

#### Features added to Secondary model
Medu/Fedu - mother's/father's education (njumeric: 0 = none, 1 = 4th grade, 2 = 5th to 9th grade, 3 = secondary education, 4 = higher education)  
reason - reason to choose this school (nominal: close to 'home', school 'reputation', 'course' preference or 'other')  
famsup - family educational support (binary: 'yes' or 'no')  
studytime - weekly study time (numeric: 1 = <2 hours, 2= 2 to 5 hours, 3 = 5 to 10 hours, 4 = >10 hours)  
activities - extra-curricular activities (binary: 'yes' or 'no')  
higher - wants to take higher education (binary: 'yes' or 'no')  
internet - internet access at home (binary: 'yes' or 'no')  
romantic - in a romantic relationship (binary: 'yes' or 'no')  
famrel - quality of family relationships (numeric: from 1 - very bad to 5 - excellent)  
freetime - free time after school (numeric: from 1 - very low to 5 - very high)  
goout - going out with friends (numeric: from 1 - very low to 5 - very high)  
Dalc - workday alcohol consumption (numeric: from 1 - very low to 5 - very high)  
Walc - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)  
health - current health status (numeric: from 1 - very bad to 5 - very good)  
  
#### Excluded Features
school - student's school. Irrelevant to a generalizable model  
Mjob/Fjob - mother's/father's job. The categorization is hazy and mostly of value 'other'  
G1 and G2 - first and second period grades. These values are too highly related to the final grade and would make the models worse.  


## Data Preparation

```python
import pandas as pd
from ucimlrepo import fetch_ucirepo   
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
# fetch dataset 
student_performance = fetch_ucirepo(id=320) 
# data (as pandas dataframes) 
X = student_performance.data.features 
y = student_performance.data.targets 

# drop unnecessary values from targets and transform target to binary
y = y.drop(columns = ['G1', 'G2'])
y_binary = (y >= 10).astype(int).values.ravel()

X = X.drop(columns = ['school', 'reason','Mjob', 'Fjob'])


# convert binaries to numerical values
X['sex'] = pd.get_dummies(X['sex'], dtype='int64')['F']
X['famsize'] = pd.get_dummies(X['famsize'], dtype='int64')['GT3']
X['Pstatus'] = pd.get_dummies(X['Pstatus'], dtype='int64')['T']
X['address'] = pd.get_dummies(X['address'], dtype='int64')['U']
X['schoolsup'] = pd.get_dummies(X['schoolsup'], dtype='int64')['yes']
X['famsup'] = pd.get_dummies(X['famsup'], dtype='int64')['yes']
X['paid'] = pd.get_dummies(X['paid'], dtype='int64')['yes']
X['activities'] = pd.get_dummies(X['activities'], dtype='int64')['yes']
X['nursery'] = pd.get_dummies(X['nursery'], dtype='int64')['yes']
X['higher'] = pd.get_dummies(X['higher'], dtype='int64')['yes']
X['internet'] = pd.get_dummies(X['internet'], dtype='int64')['yes']
X['romantic'] = pd.get_dummies(X['romantic'], dtype='int64')['yes']
X = pd.get_dummies(X, columns=['guardian'], dtype='int64', drop_first=True) 

# rename binaries by which one is 1
X.rename(columns={'sex' : 'sex_f', 'famsize':'famsize_GT3', 'Pstatus':'Pstatus_T', 'address':'address_U',}, inplace = True)

scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=512)

# Scale the training data since it's essential for the logistic regression and doesn't hurt the random forest model



X_secondary_train = X_train
X_primary_train = X_train.drop(columns = ['Medu','Fedu','famsup','studytime','activities','higher','internet','romantic','famrel','freetime','goout','Dalc','Walc','health'])
X_secondary_test = X_test
X_primary_test = X_test.drop(columns = ['Medu','Fedu','famsup','studytime','activities','higher','internet','romantic','famrel','freetime','goout','Dalc','Walc','health'])

```

# Modeling

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

forest_model_primary = RandomForestClassifier(n_estimators=100, bootstrap=True, max_samples=0.7, max_features=0.6, random_state=512)
forest_model_primary.fit(X_primary_train, y_train)

y_pred = forest_model_primary.predict(X_primary_test)
cm_forest_primary = confusion_matrix(y_test, y_pred)

print('Primary')
# visualize confusion matrix
plt.matshow(cm_forest_primary)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
for i in range(2):
    for j in range(2):
        plt.text(j, i, cm_forest_primary[i, j], ha='center', va='center', color='red')
plt.show()

print(classification_report(y_test, y_pred))
feat_importances = pd.Series(forest_model_primary.feature_importances_, index=X_primary_train.columns)
print(feat_importances.sort_values(ascending=False))


# Secondary Model

forest_model_secondary = RandomForestClassifier(n_estimators=100, bootstrap=True, max_samples=0.7, max_features=0.6, random_state=512)
forest_model_secondary.fit(X_secondary_train, y_train)

y_pred_secondary = forest_model_secondary.predict(X_secondary_test)
cm_forest_secondary = confusion_matrix(y_test, y_pred_secondary)

print('\nSecondary')
# visualize confusion matrix
plt.matshow(cm_forest_secondary)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
for i in range(2):
    for j in range(2):
        plt.text(j, i, cm_forest_secondary[i, j], ha='center', va='center', color='red')
plt.show()

print(classification_report(y_test, y_pred_secondary))
feat_importances = pd.Series(forest_model_secondary.feature_importances_, index=X_secondary_train.columns)
print(feat_importances.sort_values(ascending=False))


                                             
```
